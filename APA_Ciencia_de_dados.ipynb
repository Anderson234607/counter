```python
# Série de numero de alunos por titulação

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Criando a série
titulacao = pd.Series([20, 15, 85, 145, 320],
                      index=['Doutorado', 'Mestrado', 'Especialização', 'Graduação', 'Cursos Técnicos'])

# Exibindo a série
print(titulacao)
```

```python
# Tabela de nÚmero de alunos por titulação
import pandas as pd

# 2) Tabela que indica o número de alunos por titulação
df = pd.DataFrame({
    'Titulação': ['Doutorado', 'Mestrado', 'Especialização', 'Graduação', 'Cursos Técnicos'],
    'Quantidade': [20, 15, 85, 145, 320]
})
print("Tabela de Titulação:")
print(df)
```

```python
# Tabela Calorias e Percentual de Gordura dos Itens
import pandas as pd

# 3) Tabela que indica as calorias e o percentual de gordura de cada item
df = pd.DataFrame({
    'Calorias': [200, 350, 550],
    'Gordura (%)': [0, 15, 35]
}, index=['Banana', 'Macarrão', 'Cachorro Quente'])
print("Tabela de Alimentos:")
print(df)
```

```python
# Matriz de 50 elementos e 3 dimensões
import numpy as np

# Cria uma matriz com 50 elementos e 3 dimensões
v = np.array(range(50)).reshape(2, 5, 5)

# Exibe as informações da matriz
print('Shape = ', v.shape)
print('Número de dimensões = ', v.ndim)
print('Número de elementos = ', v.size)
print('Tensor v = \n', v)
```

```python

# Gráfico Pizza

import matplotlib.pyplot as plt

# Definindo variáveis
fatias = [8, 12, 10]
atividades = ['Lazer', 'Estudo', 'Viagem']
colunas = ['r', 'm', 'y']

# Criando um gráfico de pizza
plt.pie(
    fatias,
    labels=atividades,
    colors=colunas,
    startangle=90,
    shadow=True,
    explode=(0, 0.1, 0)  # Destaque no segundo item (Estudo)
)
plt.title("Distribuição de Atividades")
plt.show()
```

```python
#Criando Grafico em linha
import matplotlib.pyplot as plt

# Definindo variáveis
x = [1, 3, 5]
y = [1, 2, 5]

# Criando um gráfico de linha
plt.plot(x, y, label='Uma legenda')

# Atribuindo título e rótulos aos eixos
plt.title('Exemplo utilizando Plot')
plt.xlabel('Variável 1')
plt.ylabel('Variável 2')

# Exibindo a legenda
plt.legend()

# Exibindo o gráfico gerado
plt.show()
```

```python
#Criando Gráfico em estrela
import matplotlib.pyplot as plt

# Definindo variáveis
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 5, 6, 8, 4, 8]

# Criando um gráfico de dispersão com pontos em formato de estrela
plt.scatter(x, y, label='Pontos', color='b', marker='*', s=100)

# Exibindo a legenda
plt.legend()

# Exibindo o gráfico gerado
plt.show()
```
